# Simple HTTP Client using Python Sockets

This program demonstrates how to implement a basic HTTP client that:
- Establishes a TCP connection with a remote server
- Sends an HTTP GET request
- Receives and displays the webpage content

This is part of a Computer Networks project.


```python
import socket

def simple_http_get(host, path="/"):
    client_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM) # We create a TCP socket here-
    client_socket.connect((host, 80)) # Now connecting to the server

    # HTTP GET request we build here
    http_request = f"GET {path} HTTP/1.0\r\nHost: {host}\r\n\r\n"
    client_socket.send(http_request.encode())

    # Receive response data
    response = b""
    while True:
        data = client_socket.recv(4096)
        if not data:
            break
        response += data
    client_socket.close()

    # Decode response to text
    return response.decode(errors="ignore")

# For us to test :)
output = simple_http_get("example.com", "/")
print(output)

```
